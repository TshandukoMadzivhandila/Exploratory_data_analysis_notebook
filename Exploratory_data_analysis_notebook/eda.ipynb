1. Data processing

```python
import pandas as pd
import warnings
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")

data = pd.read_csv("/content/drive/MyDrive/Tutorial/data.csv")

missing_values = data.isnull().sum()
missing_values_table = pd.DataFrame(missing_values, columns=['Number of Missing Values'])
print(missing_values_table)

data = data.fillna(data.mean())

missing_values = data.isnull().sum()
missing_values_table = pd.DataFrame(missing_values, columns=['Number of Missing Values'])
print(missing_values_table)
```

2. Exploratory data analysis

```python
# Finding the country with the lowest and highest % of population under 15
lowest_under15 = data.sort_values(by='Under15').head(1)
highest_under15 = data.sort_values(by='Under15', ascending=False).head(1)

print("Country with the lowest % of population under 15:", lowest_under15["Country"].values[0])
print("Country with the highest % of population under 15:", highest_under15["Country"].values[0]+ "\n") 

# Finding the region with the highest % of population over 60
highest_over60 = data.sort_values(by='Over60', ascending=False).head(1)
print("Region with highest % of population over 60:", highest_over60["Region"].values[0]+ "\n")

# Plotting a scatter plot of FertilityRate vs GNI
fig, ax = plt.subplots()
data.plot.scatter(x='FertilityRate', y='GNI', ax=ax)
ax.set_title("Fertility Rate vs Gross National Income", fontweight='bold')
ax.set_xlabel("Fertility Rate", fontweight='bold')
ax.set_ylabel("Gross National Income", fontweight='bold')
plt.show()
print("\n")

# Finding the region with the lowest literacy rate
lowest_literacy = data.sort_values(by='LiteracyRate').head(1)
print("Region with lowest literacy rate:", lowest_literacy["Region"].values[0]+ "\n")

# Finding the region with the lowest child mortality rate
lowest_child_mortality = data.sort_values(by='ChildMortality').head(1)
print("Region with lowest child mortality rate:", lowest_child_mortality["Region"].values[0]+ "\n")

# Plotting a box plot of Life Expectancy by Region
warnings.filterwarnings("ignore")
data.boxplot(column='LifeExpectancy', by='Region')
plt.suptitle("", fontsize=12, fontweight='bold')
plt.title("Life Expectancy by Region", fontsize=12, fontweight='bold')
plt.xlabel("Region", fontsize=12, fontweight='bold')
plt.xticks(rotation=90)
plt.ylabel("Life Expectancy (in years)", fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
print("\n")

# Plotting a scatter plot of FertilityRate vs GNI
data.plot.scatter(x='FertilityRate', y='GNI')
plt.title("FertilityRate vs. GNI", fontweight='bold')
plt.xlabel("FertilityRate", fontweight='bold')
plt.ylabel("GNI", fontweight='bold')
plt.show()
print("\n")
```

3. Conclusion and Insights



*  The data set contains demographic and health information for countries in different regions. From the analysis, we can see that the regions with the lowest child mortality rates are Western Pacific and Africa. Japan and Niger, respectively, have the lowest child mortality rates among the countries in these regions. However, it's important to note that these countries have different socioeconomic conditions and that child mortality rate might be affected by other factors such as education, healthcare, and access to basic services.
*  Additionally, the analysis found that Japan and Niger have the highest and lowest life expectancy respectively. Japan has a median life expectancy of 83 years, which is higher than the median life expectancy of countries in other regions. On the other hand, Niger has a median life expectancy of 56 years, which is lower than the median life expectancy of countries in other regions. This suggests that there may be factors such as healthcare, education, and access to basic services that affect life expectancy in these countries.
